```python
import pathlib

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
print(tf.__version__)
```

```python
inputdat = pd.read_csv(r'C:\Users\yungun\Desktop\학교\2021\컴퓨터응용계통해석\IEEE30모선\input.csv')
```

```python
outputdat = pd.read_csv(r'C:\Users\yungun\Desktop\학교\2021\컴퓨터응용계통해석\IEEE30모선\output.csv')
```

```python
train_input = inputdat[0:5000]
test_input = inputdat[5000:5001]
```

```python
print(test_input )
```

```python
train_output = outputdat[0:5000] 
test_output = outputdat[5000:5001]
```

```python
input_stats = inputdat.describe()
input_stats = input_stats.transpose()
```

```python
def norm(x):
    return (x - input_stats['mean']) / input_stats['std']     #std =standard deviation 표준편차 
normed_traininput_data = norm(train_input)
normed_testinput_data = norm(test_input)
```

```python
output_stats = outputdat.describe()
output_stats = output_stats.transpose()
```

```python
def norm2(x):
    return (x - output_stats['mean']) / output_stats['std']     #std =standard deviation 표준편차 
normed_trainoutput_data = norm2(train_output)
normed_testoutput_data = norm2(test_output )
```

```python
normed_trainoutput_data
```

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(200, activation='relu', input_shape=[len(train_input.keys())]),
        
        layers.Dense(300, activation = 'relu'),
        layers.Dropout(0.5),
        
        layers.Dense(400, activation = 'relu'),
        layers.Dropout(0.5),
        
        layers.Dense(200, activation = 'relu'),
        layers.Dense(len(train_output.keys()))
        ])
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='mse', optimizer=optimizer, metrics=['mae', 'mse'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: 
            print('')            
            print('.', end='')

EPOCHS = 500

history = model.fit(
    normed_traininput_data, normed_trainoutput_data,
    epochs=EPOCHS, validation_split = 0.2, verbose=0,
    callbacks=[PrintDot()])
```

```python
test_predictions = model.predict(normed_testinput_data).flatten()
```
